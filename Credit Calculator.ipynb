# Enunciado elif.
Dado que ya estás familiarizado con los enunciados condicionales básicos, tales como los enunciados **if** y **if-else**, es momento para el más avanzado **enunciado elif**.

El **Enunciado elif** es utilizado cuando queremos especificar varias condiciones en nuetro código. ¿Cómo difiere de los **enunciados if-else**? Bien, es simple. Aquí añadimos la palabra clave **elif** para nuestras condiciones adicionales. Es utilizado en su mayoría con operadores if y else.

La condición es seguida por dos puntos :, justo como los enuncados **if-else**, entonces la acción deseada es colocada dentro del cuerpo **elif** y no hay que olvidar la sangría, es decir, los 4 espacios al principio de una nueva línea. Aquí primero verificamos la condition1 en el enunciado **if** y si se cumple (el valor de la expresión Booleana es **True**), action1 es realizada. Si es **False**, nos saltamos action1 y entonces verificamos la conition2 en el enunciado **elif**. Nuevamente, si condition2 es **true**, action2 es realizaa, de otra manera, nos lo saltamos y vamos a la parte **else** del código.

```python
# Sintaxis
if condition1:
    action1
elif condition2:
    action2
else:
    action3
    
# Es la abreviación de:
if condition1:
    action1
else:
    if condition2:
        action2
    else:
        action3
```

# Ejemplo

```python
# elif example
precio = int(input())
if precio > 5000:
    print("¡Es demasiado caro!")
elif precio > 500:
    print("¡Puedo pagar eso!")
else:
    print("¡Es demasiado barato!")
```

## Sobre el ejemplo:
¿Comprar o no comprar? Para responder a la pregunta neccesitamos ver si el precio es más grande que 5000. Si 'precio > 500' es True, imprimimos que es demasiado caro y se termina, buscando algo más barato. ¿Qué pasa si el precio es menos que 5000? En este caso, podemos verificar la siguiente condición 'precio > 500', y otra vez, si es True, imprimimos que podemos pagalo, y si es False, vamos al bloque **else** e imprimimos que es demasiado barato. Por lo tanto, '¡puedo pagar eso!' será impreso si el precio es menor que 5000 pero mayo que 500, y '¡Es demasiado barato! si el precio es menor que 500.

El enunciado **elif** difiere de **else** en un momento clave: representa otra opción específica mientras **else** encaja todos los casos que no caen en la condición dada en el enunciado **if**. Es por eso que a veces puedes encontrarte anunciados condicionales sin **else**: 

En el ejemplo de abajo, es posible añadir un enunciado **else** para expandir un poco (slightly) una perspectiva, pero no es necesario si estamos solamente interesados en perros y gatos.

```python
pet = input('mascota: ')
 
# cats vs dogs conditional
if pet == 'cat':
    print('Oh, you are a cat person. Meow!')
elif pet == 'dog':
    print('Oh, you are a dog person. Woof!')
```

# Múltiples elifs y un árbol de decisión.
Puede haber tantos enunciados **elif** como necesites, entonces tus condiciones pueden ser muy detalladas. Sin importar cuántos enunciados **elif** tengas, la sintaxis es siempre la misma. La única diferencia es que añadimos más **elifs**:

El código dentro del bloque **else** es ejecutado solo si todas las condiciones antes de él son **False**. Ve los ejemplos:

```python
# multiple elifs syntax
if condition1:
    action1
elif condition2:
    action2
# here you can add as many elifs as you need
elif conditionN:
    actionN
else:
    actionN1
```

```python
# multiple elifs example
light = "..." # some str
if light == "green":
    print("You can go!")
elif light == "yellow":
    print("Get ready!")
elif light == "red":
    print("Just wait.")
else:
    print("No such traffic light color, do whatever you want")
```

## Comentario:
En este programa, el mensaje del bloque **else** es impreso por la luz de cualquier color excepto el green, yellow y orange para los cuales hemos escrito mensajes especiales.

Los condicionales con varias ramas hacen un **árbol de decisión**, en el cual un nudo (arista) es una expresión Booleana y las ramas están marcadas con **True** o **False**. La rama marcada como True lleva a la acción que debe ser ejecutada y la rama False lleva a la siguiente condición.

# Enunciados elif anidados.
Los enunciados elif también pueden ser **anidados**, justo como los enunciados **if-else**. Podemos rescribir nuestro ejemplo de los semáforos con condiciones anidadas:

```python
# nested elifs example
traffic_lights = "green, yellow, red"
# a string with one color
light = "..."
if light in traffic_lights:
    if light == "green":
        print("You can go!")
    elif light == "yellow":
        print("Get ready!")
    else:
        # if the lights are red
        print("Just wait.")
else:
    print("No such traffic light color, do whatever you want")
```

# Literales de cadena con formato. (Formatted string literals).
**Los literales de cadena con formato** (o simplemente, **f-strings**) son utilizados para insertar los valores de expresiones dentro de literales de cadena. De esta manera se supone que es la más fácil: solo necesitas poner **f** antes de la cadena y poner las variables que quieras insertar dentro de la cadena en corchetes. También son la nueva característica entre todos los métodos de formato de cadenas en Python. Ver ejemplo 1. También puedes utilizar diferentes especificaciones de formato con f-literals, por ejemplo redondear recimales se vería así: ver ejemplo 2.

Puedes leer más información sobre el mini lenguaje de especificación de formato (Format Specification Mini-Language) en Python en la documentación oficial: https://docs.python.org/3.6/library/string.html#format-specification-mini-language

Tal vez, pensarás que estos métodos no son importantes y sobrevalorados pero ellos te darán la oportunidad de hacer tu código más atractivo y legible.

```python
# Ejemplo 1 - f-strings.
name = 'Elizabeth II'
title = 'Queen of the United Kingdom and the other Commonwealth realms'
reign = 'the longest-lived and longest-reignig British monarch'
f'{name}, the {title}, is {reign}.'

# Ejemplo 2 - redondear números.
hundred_percent_number = 1823
needed_percent = 16
needed_percent_number = hundred_percent_number * needed_percent / 100

print(f'{needed_percent}% from {hundred_percent_number} is {needed_percent_number}')

print(f'Rounding {needed_percent_number} to 1 decimal place is {needed_percent_number:.1f}')
```

# Cargar módulo.
## Básicos del módulo.

Mientras trabajaste en ejemplos simples probablemente escribiste tu código diectamente en el interpretador. Pero cada vez que cierras el interpretador y lo vuelves a abrir pierdes todas las definiciones que hiciste antes. Entonces a medida que empieces a escribir programas más grandes tiene sentido preparar tu código en adelanto utilizando un editor de texto y luego correrlo con el interpretador. Un archivo que contiene una lista de operaciones que más adelante son leídas e interpretadas es llamado **script**.

Puede que también quieras escribir algunas funciones y luego utilizarlas en otros programas o incluso reutilizar código que alguien más escribió antes. Una manera es solo copiar el código en tu programa, pero esto pronto lleva a que el código esté mal estructurado y sea difícil de leer. Por suerte, hay otra manera en Python para organizar y reutilizar código llamada **modules** módulos.

El módulo es simplemente un archivo que contiene enunciados de Python y definiciones. Usualmente tiene una extensión **.py**. Lo que realmente hace poderoso el sistema de módulos es la capacidad para **cargar** o **importar** un módulo desde otro. 

## Cargando un módulo.

Para cargar un módulo solo usa un enunciado **import**. De manera básica, tiene la siguiente sintaxis: **import module**, donde module es el nombre del módulo que se quiere cargar. Ver ejemplo 1.

**super_module** es el nombre del módulo que quieres importar. Por ejemplo, un archivo llamado **super_module.py** tiene un nombre **super_module**. Para ser capaces de importar, **super_module.py** debe ser colocado en el mismo directorio del archivo en el que estás intentando importarlo. Al inicio, el sistema de importación de Python busca un módulo en el directorio actual, luego verifica los módulos predefinidos (built-in), y si no encuentra nada un error será levantado. Después de importar, el módulo se vuelve disponible bajo su nombre y puedes acceder a funciones y variables definidas en él utilizando la notación punto. 

También es común solamente importar funciones o variables requeridas desde un módulo pero no el módulo por sí mismo. Puedes hacer esto utilizando una forma **from** del enunciado import, ver ejemplo 2.

Una buena práctica es cargar un solo módulo en una sola línea y poner todos las importaciones _al principio del archivo_ porque esto mejor su legibilidad, ver ejemplo 3.

Una forma especial del enunciado import permite cargar todos los nombres definidos en un módulo. Se llama _wildcart import_ y tiene sintaxis: **from module import*** (debe ir separado: import *) Generalmente debes evitar esto en tu código. Puede causar comportamiento inesperado porque no sabes qué nombres exactamente son importados en el actual espacio del nombre (*). Por otra parte, estos nombres pueden hacer al lado algunos de los existentes sin que sepas. Es mejor hacerlo explícitamente y especificar lo que estás importando. 

En caso de que tengas que utilizar varios enunciados import, pon atención a su orden:

1. imports de librerías estándar.
2. imports de dependencias terciarias.
3. import de aplicaciones locales.

Tener tus importaciones agrupadas, puedes poner una línea en blanco entre las sección de import. Además, algunas guías, incluyendo la nuestra, recomiendan ordenar las importaciones de manera alfabética.

1. import super_module <- importa todo el módulo.
2. from super_module import super_function <- importar una sola función desde el módulo.
3. from module import * <- importar todos los nombres del módulo.
4. super_module.super_function() <- acceso a la función cargada desde el módulo.

La diferencia entre importar el nombre de todas las funciones/variables e importar el módulo es la manera en que accedes a las funciones/variables.

```python
# Ejemplo 1
import super_module

super_module.super_function()  # llamando a una función definida en super_module

print(super_module.super_variable)  # accediendo a una varible definida en super_module

# Ejemplo 2
from super_module import super_function

super_function()  # super_function() ahora está disponible directamente en el modulo actual

super_module.super_function()  # notar, en este caso super_module no está importado,
                               # por lo tanto esta línea genera un error.     

# Ejemplo 3
import module1
import module2
import module3

# el resto del código del módulo va aquí.
```

## Aritmética avanzada.

Hay funciones predefinidas abs, round, pow, max y min: (ver ejemplo 1)

+ **abs(x)** regresa el valor absoluto de **x**(es decir, el valor sin considerar su signo);

+ **max(a, b, c,...)** regresa el argumento mayor;

Las funciones **abs** y **pow** tienen equivalencias en el módulo **math**. La diferencia clave de math.fabs y math.pow es que siempre regresan floats, ver ejemplo 2.

# El interpretador.

Generalmente, ¿qué es un proceso de interpretación? Uno puede decir que se parece a **leer** un programa. Algunos softwares solo "leen" tu programa y ejecutan lo que está escrito en él **línea por línea**. Este software es inesperadamente llamado **interpretador** y es parte del paquete de instalación estándar de Python. Otra parte inherente es la **librería estándar** con todos los módulos predeterminados (built-in), funciones, datos, etc. Un hecho interesante: un interpretador puede ser escrito en básicamente cualquier lenguaje de programación. El interpretador predeterminado para Python está escrito en C y es llamado **CPython**. Algunos otros interpretadores son: 

+ **PyPy** está escrito en un subconjunto restringido de Python llamado RPython (Python restringido), el cual provee algunas restricciones al código usual de Python.
+ **Jython** traduce el código de Python in código byte compatible con Java, el cual es más tarde ejecutado por JVM - Java Virtual Machine.
+ **IronPython** es una interpretación de Python para el .NET framework.

Ahora intentemos entender, lo que está realmente ocurriendo durante la interpretación. De hecho, este paso consiste de 3 más pequeños:

---------------------------------Interpretación------------------------
Escribiendo un script ----> Compilación | Bytecode | Virtual Machine  -----> ¡Todo funciona!

El compilador convierte tu conjunto de enunciados (código fuente) en los llamados **códigos byte**. El código byte es por sí mismo una versión de código fuente de **bajo nivel** (por lo tanto más detallado), **independiente de plataforma** y **más eficiente**, pero no es **código binario de máquina** como instrucciones para un chip Intel o AMD. El código byte es una representación **especifica de Python** de código fuente. Es por eso que algunos programas de Python son ejecutados no tan rápido como los análogos en C++ o C - lenguajes compilados tradicionales.

# Python Virtual Machine.

Después de la compilación, el código byte es dado en el **PVM (Python Virtual Machine)**. A pesar de que suena algo impresionante, de hecho, no es más que una gran pieza de código, que itera a través de intrucciones de código byte recibidas desde un compilador y las ejecuta **una por una**, por lo tanto realiza operaciones deseables especificadas por el programador. Es una parte interna del sistema de Python y no ncecesitas instalarlo dado que no es un programa separado. De hecho, la cosas que realmente ejecuta es PVM, entonces podemos decir que es el último paso de ejecución de cualquier programa de Python. Toda esta complejidad es deliberadamente escondida de un programador. La "parte de interpretación" es **totalmente automatizada**, entonces usualmente no necesitas pensar en ello. Recordad: los programadores de Python simplemente escriben el código y corren los archivos, todo lo demás es hecho por Python.

# Conclusion.

Ahora puedes entender que sí, Python en efecto es un lenguaje interpretado, pero antes de la etapa de interpretación, hay un proceso interno de compilar el código fuente a código byte. Por lo tanto, ejecutar un programa de Python implica ambos pasos: compilación e interpretación.

Cuando tu programa es convertido en código byte, es ejecutado **línea por línea desde arriba hasta abajo**, entonces generalmente puedes esperar que tu programa funcione exactamente en la manera que lo has escrito.

## ¿Qué es una iteración?

Las computadoras son conocidas por su habilidad para hacer cosas las cuales las personas considran aburridas y consumidoras de energía. Por ejemplo, repetir tareas idénticas sin errores es una de esas cosas. En Python, el proceso de ejecutión repetitiva del mismo bloque de código es llamado una **iteración**.

Hay dos tipos de iteración:

+ **Iteración definida**, donde el número de repeticiones es establecido de antemano.

+ **Iteración indefinida**, donde el bloque de código se ejecuta siempre y cuando la condición establecida de antemano es verdadera.

Después de la primera iteración, el programa regresa al inicio del cuerpo del código y lo repite, haciendo el tan famoso **loop**. El más utilizado es el **loop for**, nombrado después del **operador for** que provee la ejecución del código.

# Sistema operativo.

**Un sistema operativo (OS)** es un conjunto de software que maneja la comunicación entre todas las otras aplicaciones y el hardware. Vuelve a una computadora en algo más que solo un montón de piezas de metal, un sistema complejo que puede realizar efectivamente diferentes tareas.

# Funciones del OS.

Un sistema operativo controla la comunicación entre todo el software de la computadora y el hardware. Un OS puede dar acceso restringido a programas para procesar unidades, memoria, discos duros, red, periféricos (peripherals) y otros recursos.

Puedes correr un buscador, un reproductor de video, y otras diez aplicaciones, y tu OS es el que les da todos los recursos que necesitan para correr adecuadamente. Al mismo tiempo, este OS actúa como un referi prohibiendo que alguna aplicación tome más recursos de los que realmente necesita.

Como un mediador entre las aplicaciones y el hardware, el sistema operativo permite que nos comuniquemos con el dispositivo sin entrar en detalles sobre sus especificaciones o mecánicas.

# Módulo del sistema

No entraremos endetalles sobre sus características, más bien nos enfocaremos en una de las cosas más importante ahora, llamado, **sys.argv**. Realiza la operación que necesitamos: colectar los argumentos pasados al script de Python.

Llamando **sys.argv**, obtenemos los argumentos especificados por el usuario como **una lista de cadenas**. Indexando, como siempre en Python, empiza desde 0 el primer elemento, **sys.argv[0]** es el nombre de nuestro script de Python como fue llamado - el nombre por sí mismo o incluyendo la dirección del archivo. Y los siguientes elementos de la lista son los argumentos que podemos acceder a ellos por su índice. Toma nota que son cadenas, y si necesitamos un valor numérico, tenemos qu realizar la conversión de tipo.

```python
import math

print('What do you want to calculate?')
print('type "n" - for count of months,')
print('type "a" - for annuity monthly payment,')
print('type "p" - for credit principal')
choice = input()

if choice == 'n':
    print('Enter credit principal:')
    credit_principal = int(input())
    print('Enter monthly payment:')
    monthly_payment = int(input())
    print('Enter credit interest:')
    credit_interest = input()
    i = (float(credit_interest) / 100) / 12 
    n = math.ceil(math.log(monthly_payment / (monthly_payment - i * credit_principal), 1+ i))
    years =  n // 12
    months = n % 12
    if years == 0:
        if months == 1:
            print(f'You need {months} month to repay this credit!')
        else:
            print(f'You need {months} months to repay this credit!')
    elif months == 0:
        if years == 1:
            print(f'You need {years} year to repay this credit!')
        else:
            print(f'You need {years} years to repay this credit!')
    else:
        print(f'You need {years} years and {months} to repay this credit!')
elif choice == 'a':
    print('Enter credit principal:')
    credit_principal = int(input())
    print('Enter count of periods:')                          
    count_periods = int(input()) 
    print('Enter credit interest:')
    credit_interest = input() 
    i = (float(credit_interest) / 100) / 12 
    annuity = math.ceil(credit_principal * (i * pow(i + 1, count_periods)) / (pow(1 + i, count_periods) - 1))
    print(f'Your annuity payment = {annuity}!')
elif choice == 'p':
    print('Enter monthly payment:')
    monthly_payment = float(input())
    print('Enter count of periods:')                          
    count_periods = int(input()) 
    print('Enter credit interest:')
    credit_interest = input()
    i = (float(credit_interest) / 100) / 12 
    credit_principal = math.floor(monthly_payment / (i * pow(1 + i, count_periods) / (pow(1 + i, count_periods) - 1)))
    print(f'Your credit principal = {credit_principal}!')
```

```python
import argparse
import math
import sys

parser = argparse.ArgumentParser()
parser.add_argument("--type")
parser.add_argument("--payment")
parser.add_argument("--principal")
parser.add_argument("--periods")
parser.add_argument("--interest")

args = parser.parse_args()


if len(sys.argv) < 5:
    print('Incorrect parameters')
elif args.type == 'annuity':
    # annuity
    if args.payment == None:
        i = (float(args.interest) / 100) / 12
        annuity = math.ceil(float(args.principal) * (i * pow(i + 1, float(args.periods))) / (pow(1 + i, float(args.periods)) - 1))
        print(f'Your annuity payment = {annuity}!')
        overpayment = annuity * int(args.periods) - int(args.principal)
        print(f'Overpayment = {overpayment}')
    # credit principal
    elif args.principal == None:
        i = (float(args.interest) / 100) / 12
        credit_principal = math.floor(float(args.payment) / (i * pow(1 + i, float(args.periods)) /
                                                             (pow(1 + i, float(args.periods)) - 1)))
        print(f'Your credit principal = {credit_principal}!')
        overpayment = int(args.payment) * int(args.periods) - credit_principal
    # periods
    elif args.periods == None:
        i = (float(args.interest) / 100) / 12
        n = math.ceil(math.log(float(args.payment) / (float(args.payment) - i * float(args.principal)), 1 + i))
        years = n // 12
        months = n % 12
        if years == 0:
            if months == 1:
                print(f'You need {months} month to repay this credit!')
            else:
                print(f'You need {months} months to repay this credit!')
        elif months == 0:
            if years == 1:
                print(f'You need {years} year to repay this credit!')
            else:
                print(f'You need {years} years to repay this credit!')
        else:
            print(f'You need {years} years and {months} to repay this credit!')
        overpayment = (years * 12 + months) * int(args.payment) - int(args.principal)
        print(f'Overpayment = {overpayment}')
elif args.type == 'diff':
    if args.payment != None:
        print('Incorrect parameters')
    else:
        i = (float(args.interest) / 100) / 12
        counter = 0
        for m in range(1, int(args.periods) + 1):
            d = math.ceil(float(args.principal) / float(args.periods) + i * (float(args.principal) - float(args.principal) * (m - 1)
                                                                   / float(args.periods)))
            print(f'Month {m}: paid out {d}')
            counter += d
        overpayment = counter - int(args.principal)
        print(f'Overpayment = {overpayment}')
else:
    print('Incorrect parameters')
```
